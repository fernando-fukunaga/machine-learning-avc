# ðŸŽ² TRATAMENTO DE DADOS

```python
import pandas as pd

stroke_data_set = pd.read_csv('dados1.csv', delimiter=';')

stroke_data_set.shape
stroke_data_set.head(10)
```

```python
stroke_data_set.describe()
```

```python
avc_negativo = stroke_data_set.loc[stroke_data_set['AVC'] == 0]

avc_negativo_249 = avc_negativo.sample(n=249)

avc_negativo_249
```

```python
avc_positivo = stroke_data_set.loc[stroke_data_set['AVC'] == 1]

avc_positivo
```

```python
final_dataset = pd.concat([avc_positivo, avc_negativo_249], ignore_index=True)

final_dataset_shuffle = final_dataset.sample(frac=1).reset_index(drop=True)

final_dataset_shuffle
```

# ðŸ¤– MODELO E TREINAMENTO

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
final_dataset_shuffle_target = final_dataset_shuffle[['AVC']]
final_dataset_shuffle_data = final_dataset_shuffle.drop('AVC', axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(final_dataset_shuffle_data,
                                                    final_dataset_shuffle_target,
                                                    test_size=0.25)
```

```python
tree_ia = tree.DecisionTreeClassifier(criterion='entropy', max_depth=5)
```

```python
tree_ia = tree_ia.fit(x_train, y_train)
```

```python
predictions = tree_ia.predict(x_test)
```

```python
print(predictions)
```

```python
y_test_series = y_test.squeeze()
predictions_series = pd.Series(predictions)
```

```python
matriz = pd.crosstab(y_test_series,
                     predictions_series,
                     rownames=["Real"],
                     colnames=["Predito"],
                     margins=True,
                     margins_name='Todos')

print(f"Matriz de confusÃ£o detalhada:\n\n{matriz}")

acertos_1 = matriz.loc[0,0]
acertos_2 = matriz.loc[1,1]
todos = matriz.loc['Todos','Todos']

acertos_total = acertos_1 + acertos_2
acuracia = acertos_total / todos

acuracia_formatada = round(acuracia * 100, 1)
acuracia_formatada

print(f"\nAcurÃ¡cia de aproximadamente {acuracia_formatada}%")
```
